# Parts of a Figure

The matplotlib graphs contain many different parts of a figure that we should know how to interact with and what they refer to.

### Figure

The whole figure - keeps track of all the child **Axes**, also can be refered to as a smattering of 'special artists (title, figure, legends ect.) and the **Canvas** (don't worry too much about the Canvas, it is a crucial as it is the object that does the drawing to get you the plot).

A Figure can contain any number of **Axes** but will typically have at least one.

The easiest way to create a new figure is with pyplot:

```python
fig = plt.figure() # a Figure with no Axes
ax = plt.subplot() # a figure with a single Axes
fig,ax = plt.subplots(2,2) #a Figure with 2x2 grid of Axes
```

It is convenient to create the axes together with the figure, but you can add axes later on, allowing for more complex axes and layouts.

### Axes

This is what you think of as a 'plot', it is the region of the image with the data space. A given figure can contain many Axes, but a given Axes object can only be in one figure.

The **Axes** object contains two (or Three) **Axis** objects.

It is important to know the differences between the Axes and Axis object.

The data limits are controlled by the Axes and can be set with the following:

*axes.Axes.set_xlim()* and *axes.Axes.set_ylim()* methods. Each Axes has a title (*set_title()*), an x-label (*set_xlabel*) and a y-label (*set_ylabel()*).

The **Axes** class and its members functions are the primary entry point to working with the OO interface.

### Axis

These are the number-line-like objects. They take care of setting the graph limits and generating the ticks (marks on the axis) and ticklabels.

The location of the ticks is determined by a locator object and the tick labels strings are formated by a **Formatter**.

The combination of a **Locator** and **Formatter** gives very fine control over the tick locations and labels.

### Artist

Everything you can see on the figure is an artist (even the **Figure** , **Axes** and **Axis**)

# Types of inputs to plotting functions

All of the plotting functions accept **numpy.array** or **numpy.ma.masked_array** as input. Classes that are 'array-like' such as **pandas** data objects and **numpy.matrix** may or may not work as intended.

### Interacting with all of the elements

Since all of the elements of a Plot is it's own object this becomes a powerful too when we know how to interact with these elements. We can look at a
Example starting with creating the first few elements

## Example

We will start by building up a basic plot before we start the customization or seeing how we can interact with the objects.py

We will first create a single Figure with a single axes.

```python
fig, ax = plt.subplots()
```

We will now add some data into the figure in order to have something to add to it (This will be done random data).

```python
fig, ax = plt.subplots()

# Create the data to plot
X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

ax.scatter(X, Y3)
ax.plot(X, Y1)
ax.plot(X, Y2)
```

As we can see we now have a colourful plot that may be slightly small and requires some text labels

One of the benefits of Matplotlib creating the elements as different objects it lets us interact with their components.py

We can first check the methods avaliable to us with the code below (We will do it with the figure for this example)

```python
dir(fig)
```

From this list there are a few get elements. Let's pretend that we did not use the **subplots** method to create the figure
and want to get the axes on the figure:

```python
axes_list = fig.get_axes()
axes_list
```

We can now see all of the Axes associated with the Figure. We get get the individual axes with the following code now:

```python
ax_ex = axes_list[0]
ax_ex
```

Now that we understand that we can interact with all of the properties of the objects of Matplotlib and the list them
we can start working on improving the previous plot we made.

We will now use the set parameter function or **setp** for the figure to set up some of the elements:

```python
plt.setp(fig)
```

Since we did not include the parameter to modify we have the list of parameters returned to us with the call.

Let's change the size_inches to improve the size:

```python
plt.setp(fig,size_inches=(10,10))
fig
```

Now that is looking better in terms of the size of the plot. Let's now make changes to the axes:

First we can explore the properties we can change possibly:

```python
plt.setp(ax)
```

```python
plt.setp(
    ax,
    xlabel="Simple x range",
    ylabel="Simply y range",
    title="Meaningless jumble of couple of cosines",
    xlim=(0, 4),
    ylim=(0, 4.5),
)
fig
```

Before we modify the graph elements we need to discuss *Containers* and *Primatives*.

These two items are two classes that are essential to how we work with Matplotlib. An example of containers is the Figures
and axes while primatives are all the graphical or geometrical objects that go into a container.

We can access all of the names of the objects under the patches object of Matplotlib:

```python
from matplotlib import patches
dir(patches)
```

From this we have every thing to mkae any kind of graph. We can now look at all of the objects that are
contained in the axes:

```python
ax.get_children()
```

This shows us the twolines under Line2d and pathcollection for the scatter plot.py

Let's modify the scatter plot now using this information:

```python
plt.setp(ax.get_children()[0], edgecolor="red", facecolor="red", sizes=[40])

fig
```
